```python
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
df["isfemale"]= [1 if (x =="female") else 0 for x in df["gender"]]
```

```python
df.head()
```

```python
#No gender

import statsmodels.formula.api as sm
result = sm.ols(
formula='grade ~ age + exercise + hours',
data=df).fit()
result.summary()
```

```python
#Gender

import statsmodels.formula.api as sm
result = sm.ols(
formula='grade ~ age + exercise + hours + isfemale',
data=df).fit()
result.summary()
```
